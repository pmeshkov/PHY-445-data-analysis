```python
import numpy
import matplotlib.pyplot as plt
import math
import random
import scipy
```

```python
arrayx = numpy.arange(0.1,100,1)
```

```python
def sincrand(x):
    return 2.5*math.sin(0.5*x)/x + (random.uniform(0, 1)-0.5)*0.15

vectorsincrand = numpy.vectorize(sincrand)
```

```python
arrayy = vectorsincrand(arrayx)
```

```python
plt.plot(arrayx, arrayy)
```

```python
def sinc(x, a, w):
    return a*math.sin(w*x)/x
vectorsinc = numpy.vectorize(sinc)
```

```python
parameters, covariance = scipy.optimize.curve_fit(vectorsinc, arrayx, arrayy, p0 = [2,0.75])
a, w = parameters
print(a,w)
```

```python
arrayyfit = vectorsinc(arrayx, a, w)
```

```python
errorarray = [0.03 for elem in arrayx]
```

```python
plt.clf()
fig = plt.figure()
ax = fig.add_subplot()

ax.errorbar(arrayx, arrayy, yerr=errorarray, label = "Experimental Data", color = "orange", alpha = 0.6, linestyle='', marker='.')
ax.plot(arrayx, arrayyfit, label = "Curve Fit")

ax.set_xlabel("Angle [radians]")
ax.set_ylabel("Amplitude")

ax.set_title("Amplitude vs Angle")

ax.legend()
```
